```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
a=loadtxt('leapfrog_norm.out').T
```

```python
plot(a[0],a[2])
```

```python
from matplotlib import animation
fig, ax = plt.subplots(figsize=(5,5))
p, = ax.plot(a[0],a[1])

ax.set_xlim([1,2])
ax.set_ylim([0,1])

def init():
    p.set_data(a[0],a[2])
def update(n):
    p.set_data(a[0],a[len(a)//frames*n])
frames=150
anim = animation.FuncAnimation(fig, update, init_func=init, 
        frames=frames, blit=True)
anim.save('animation.mp4', fps=20);
```

```python
from IPython.display import HTML
video = open("animation.mp4", "rb").read()
video_encoded = video.encode("base64")
video_tag = '<video controls alt="test" src="data:video/mp4;base64,{0}">'.format(video_encoded)
HTML(video_tag)
```
